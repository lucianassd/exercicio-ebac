```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %matplotlib notebook
```

#### Leitura do arquivo

```python
sinasc = pd.read_csv('sinasc_RO_2019.csv')
sinasc.head(2)
```

#### Colunas do dataframe

```python
sinasc.columns
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sinasc['IDADEMAE'].dtypes
```

```python
box_idade = sns.boxplot(y='IDADEMAE', data=sinasc)
box_idade.set_ylabel('Idade em anos')
box_idade.set_title('Idade da mãe')
```

####  2) Faça um boxplot do peso do bebê

```python
sinasc['PESO'].dtypes
```

```python
box_peso = sns.boxplot(y='PESO', data=sinasc)
box_peso.set_ylabel('Peso em gramas')
box_peso.set_title('Peso do bebê ao nascer')
```

####  3) Faça um histograma da idade da mãe

```python
hist_idade = sns.displot(data=sinasc, x='IDADEMAE', bins=10)
hist_idade.set_xlabels('Idade em anos')
plt.title('Idade da mãe')
```

#### 4) Faça um histograma do peso do bebê

```python
hist_peso = sns.displot(data=sinasc, x='PESO', bins=12)
hist_peso.set_xlabels('Peso em gramas')
plt.title('Peso do bebê ao nascer')
```
