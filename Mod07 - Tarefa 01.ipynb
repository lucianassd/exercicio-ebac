####  Criação do novo DataFrame

````python
meta_dados = pd.DataFrame(columns = ['variavel', 'tipo_dado'])
meta_dados
````

````python
for col in df_credit.columns:
    meta_dados = pd.concat([meta_dados, pd.DataFrame({'variavel': [col], 'tipo_dado': [df_credit[col].dtype]})])
````

#### Contagem da quantidade de categorias 

````python
contagem = []

for col in df_credit:
    cont = df_credit[col].nunique()
    contagem.append(cont)
contagem
````

#### Criação da nova coluna e atribuição dos valores de contagem

````python
meta_dados['qtd_categorias'] = contagem
meta_dados[['qtd_categorias']]
````

````python
meta_dados
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
df_dummy = pd.get_dummies(df_credit, columns=['tipo_renda', 'educacao', 'estado_civil'], drop_first=True)
df_dummy
````

#### Transformação dos tipo de dados

````python
df_dummy['sexo'].unique()
````

````python
df_dummy.sexo = df_dummy.sexo.map({'M': 1, 'F':0})
````

````python
df_dummy['posse_de_veiculo'].unique()
````

````python
df_dummy.posse_de_veiculo = df_dummy.posse_de_veiculo.map({'Y': 1, 'N':0})
````

````python
df_dummy['posse_de_imovel'].unique()
````

````python
df_dummy.posse_de_imovel = df_dummy.posse_de_imovel.map({'Y': 1, 'N':0})
````

````python
df_dummy = df_dummy.drop(['qtd_filhos', 'tipo_residencia', 'qt_pessoas_residencia'], axis=1)
df_dummy.shape
````

````python
df_dummy.dtypes
````

````python
df_dummy.mau.astype(int)
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
freq = pd.crosstab(df_dummy['possui_email'], df_dummy['posse_de_veiculo'])
freq
````

#### Resposta

````python
A variável posse_de_veiculo me parece mais poderosa porque a frequencia é muito maior em relação a possui_email.
````
